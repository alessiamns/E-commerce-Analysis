```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_payments = pd.read_csv('olist_order_payments_dataset.csv')
```

## Pie Plot

```python
a = df_payments.groupby(['payment_type']).count()
a = a.reset_index()
list(a['payment_type'])
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
explode = (0, 0, 0, 0, 0.1)
col = list(a['payment_type'])
orders = list(a['order_id'])
ax.pie(orders, explode=explode, labels=col, autopct='%1.2f%%', startangle=90)
plt.show()
```

# Istogramma

```python
# Example data
listp = list(a['payment_type'])
labels = listp
count = list(a['order_id'])
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(8,8))

ax.bar(labels, count, width, label='Pay')
ax.set_title('Numero di ordini suddivisi per tipologia di pagamento')
ax.set_xlabel('Payment Type')
ax.set_ylabel('Orders')
ax.legend()
plt.show()
```

### Soglie pagamenti

```python
df_payments['payment_value'].describe()
```

```python
df_0_25 = df_payments[df_payments['payment_value'] < 56.79].groupby(['payment_type']).count()
df_0_25.sort_values(by='order_id', ascending=False)
```

```python
df_25_50 = df_payments[(df_payments['payment_value'] > 56.79) & (df_payments['payment_value'] < 100)].groupby(['payment_type']).count()
df_25_50.sort_values(by='order_id', ascending=False)
```

```python
df_50_100 = df_payments[(df_payments['payment_value'] > 100) & (df_payments['payment_value'] < 171.8375)].groupby(['payment_type']).count()
df_50_100.sort_values(by='order_id', ascending=False)
```

```python
df_100_max = df_payments[(df_payments['payment_value'] > 171.8375) & (df_payments['payment_value'] < 13664.08)].groupby(['payment_type']).count()
df_100_max.sort_values(by='order_id', ascending=False)
```
