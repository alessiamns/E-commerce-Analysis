```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure as fg
import numpy as np
```

# Products

```python
df_products = pd.read_csv('olist_products_dataset.csv')
```

```python
df_products.head()
```

### Group by category name

```python
df_category = df_products.groupby(['product_category_name']).count()
```

```python
df_category = df_category.sort_values(by='product_id',ascending=False).head(20)
df_category = df_category.reset_index()
df_category
```

```python
#df_category = df_category.loc[:, ['product_category_name', 'product_id']]
#df_category = df_category.rename(columns={"product_id": "count"})
df_category
```

## Bar Plot - Categories

```python
#fg(num=None, figsize=(12, 12), dpi=80, facecolor='w', edgecolor='k')

fig, ax = plt.subplots(figsize=(8,8))

# Example data
category = list(df_category['product_category_name'])
y_pos = np.arange(len(category))
x_pos = list(df_category['count'])

ax.barh(y_pos, x_pos, color='green')
ax.set_yticks(y_pos)
ax.set_yticklabels(category)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Categories')
ax.set_title('First 20 categories')
#plt.show()
fig.savefig('20_categories.png')
```
